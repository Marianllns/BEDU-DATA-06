## Analisis de datos 

Saber cuales de los diferentes sitios web (sources) tienen mpas interacción

```python
# sELECCIONAR LAS COLUMNAS BASE PARA EL ANALISIS

COLUMNS = [
    'source_name',
    'author',
    'engagement_reaction_count',
    'engagement_comment_count',
    'engagement_share_count',
    'engagement_comment_plugin_count'
]

#lIMPIAR TODAS AQUELLAS FILAS QUE TIENEN UN CALOR NULO EN LAS COLUMNAS SELECCIONADAS
data = data[COLUMNS].dropna(axis=0, how='any')
```

```python
#Agrupamos nuestro df por la columna source_name y sumamos cada fila por su 'engagement_reaction_count'

engagement_reaction_count = data.groupby('source_name') ['engagement_reaction_count'].sum()

engagement_reaction_count.to_frame()
```

```python
#Agrupamos nuestro df por la columna source_name y sumamos cada fila por su 'engagement_share_count'

engagement_share_count = data.groupby('source_name') ['engagement_share_count'].sum()

engagement_share_count.to_frame()
```

```python
# extraer nueva serie

highest_source = data['source_name'].value_counts()
type(highest_source)
highest_source.to_frame()
```

### Visualización de datos

```python
#seleccionamos el valor de la primera columna (indice) y lo convertimos a un elemento de lista por cada valor en la serie

labels = list(highest_source.index.values)
print(labels)
```

```python
# Crear gráfica de pastel

plt.pie(highest_source, labels=labels, autopct = '%1.1f%%')
plt.title('Distribución de artículos por fuente')

plt.show()
```
